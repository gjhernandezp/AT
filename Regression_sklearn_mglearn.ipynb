```
import mglearn
from sklearn.neighbors import KNeighborsRegressor

import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)
model = KNeighborsRegressor().fit(X,y)
x_test= np.linspace(-4,4, 1000)
y_pred =model.predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()
```

```
import mglearn
from  sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=30)
model = LinearRegression().fit(X, y)
x_test= np.linspace(-4,4, 1000)
y_pred =model.predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()
```

```
import mglearn
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)
model = MLPRegressor(random_state=1, max_iter=500).fit(X,y)
x_test= np.linspace(-4,4, 1000)
y_pred =model .predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()
```

```
import mglearn
from sklearn.tree import DecisionTreeRegressor 
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)
names = ["x"]
rt = model = DecisionTreeRegressor(max_leaf_nodes = 10).fit(X,y)
x_test= np.linspace(-4,4, 1000)
y_pred =model.predict(x_test.reshape(-1, 1))
plt.plot(X,y,'.', color='blue')
plt.plot(x_test, y_pred, color='red')
plt.show()

```

```
import graphviz # for plotting decision tree graphs
from sklearn import tree 

dot_data = tree.export_graphviz(rt, out_file=None, 
                                feature_names=["x"], 
                                filled=True, 
                                rounded=True, 
                                #rotate=True,
                               ) 

graph = graphviz.Source(dot_data)

graph
```
